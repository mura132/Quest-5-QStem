**График scatterplot пытается показать зависимость между количеством дней на работе (days employed) и возрастом человека (DAYS_BIRTH) для всех, кроме безработных. Обратите внимание, что для этого графика Вам нужно будет сначала удалить безработных людей из набора данных.
```python
# НАПИШИТЕ ВАШ КОД В ЭТОЙ ЯЧЕЙКЕ
```

```python
df_employed = df[df['DAYS_EMPLOYED'] < 0]
df_employed['DAYS_EMPLOYED'] = -df_employed['DAYS_EMPLOYED']
df_employed['DAYS_BIRTH'] = -df_employed['DAYS_BIRTH']
plt.figure(figsize=(9, 6))
plt.scatter(df_employed['DAYS_BIRTH'], df_employed['DAYS_EMPLOYED'], alpha=0.1, linewidths=0.5)
plt.title('Scatter Plot: Days Employed vs Days Birth')
plt.xlabel('Days Birth')
plt.ylabel('Days Employed')
plt.show()
```

### ЗАДАНИЕ: воссоздайте приведённый ниже график Categorical:

<img src='catplot_solution.png'>

**Этот график отображает информацию только для *половины всех людей*, имеющих доход в этом наборе данных. Причём мы берём "нижнюю" половину - тех, кто зарабатывает меньше. Графики boxplot отображены для каждого значения колонки NAME_FAMILY_STATUS, показывая распределения их общего дохода. Раскраска hue сделана по колонке "FLAG_OWN_REALTY". Замечание: чтобы взять только нижнюю половину распределения дохода, Вам нужно будет преобразовать dataframe или взять только его часть *перед* тем, как рисовать график.**

```python
# НАПИШИТЕ ВАШ КОД В ЭТОЙ ЯЧЕЙКЕ
df_lower_half_income = df[df['AMT_INCOME_TOTAL'] <= df['AMT_INCOME_TOTAL'].median()]

sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))
sns.boxplot(x='NAME_FAMILY_STATUS', y='AMT_INCOME_TOTAL', hue='FLAG_OWN_REALTY', data=df_lower_half_income)
plt.title('Boxplot: Total Income Distribution by Family Status')
plt.xlabel('NAME_FAMILY_STATUS')
plt.ylabel('AMT_INCOME_TOTAL')
plt.legend(title='Flag_Own_Reality',loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.show()
```
