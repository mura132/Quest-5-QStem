```python
from collections import deque

class Graph:
    def __init__(self):
        self.adj_list = {}

    def add_vertex(self, vertex):
        if vertex not in self.adj_list:
            self.adj_list[vertex] = []

    def add_edge(self, vertex1, vertex2):
        self.add_vertex(vertex1)
        self.add_vertex(vertex2)
        self.adj_list[vertex1].append(vertex2)
        self.adj_list[vertex2].append(vertex1)  # Uncomment for undirected graph

    def bfs(self, start_vertex):
        visited = set()
        queue = deque([start_vertex])
        result = []

        while queue:
            current_vertex = queue.popleft()
            if current_vertex not in visited:
                visited.add(current_vertex)
                result.append(current_vertex)
                queue.extend(self.adj_list[current_vertex])

        return result

# 3)Write a program for breadth-first search in a graph.
# Creating a graph
graph = Graph()

# Adding vertices and edges
graph.add_edge('A', 'B')
graph.add_edge('A', 'C')
graph.add_edge('B', 'D')
graph.add_edge('B', 'E')
graph.add_edge('C', 'F')

# Performing BFS starting from vertex 'A'
result = graph.bfs('A')
print("BFS Result:", result)
```

```python
import heapq

class Graph:
    def __init__(self):
        self.adj_list = {}

    def add_vertex(self, vertex):
        if vertex not in self.adj_list:
            self.adj_list[vertex] = {}

    def add_edge(self, vertex1, vertex2, weight):
        self.add_vertex(vertex1)
        self.add_vertex(vertex2)
        self.adj_list[vertex1][vertex2] = weight
        self.adj_list[vertex2][vertex1] = weight  # Uncomment for undirected graph

    def shortest_path(self, start_vertex, end_vertex):
        heap = [(0, start_vertex)]
        visited = set()

        while heap:
            current_weight, current_vertex = heapq.heappop(heap)

            if current_vertex in visited:
                continue

            visited.add(current_vertex)

            if current_vertex == end_vertex:
                return current_weight

            for neighbor, weight in self.adj_list[current_vertex].items():
                if neighbor not in visited:
                    heapq.heappush(heap, (current_weight + weight, neighbor))

        # If the end vertex is not reachable from the start vertex
        return float('inf')

# 1)Write a program to find the shortest path between two vertices in a graph.
# Creating a graph
graph = Graph()
graph.add_edge('A', 'B', 1)
graph.add_edge('A', 'C', 4)
graph.add_edge('B', 'C', 2)
graph.add_edge('B', 'D', 5)
graph.add_edge('C', 'D', 1)
graph.add_edge('D', 'E', 7)

# Finding the shortest path from 'A' to 'E'
start_vertex = 'A'
end_vertex = 'E'
result = graph.shortest_path(start_vertex, end_vertex)

if result == float('inf'):
    print(f"There is no path from {start_vertex} to {end_vertex}.")
else:
    print(f"Shortest path from {start_vertex} to {end_vertex}: {result}")
```

```python
import heapq

class Graph:
    def __init__(self):
        self.adj_list = {}

    def add_vertex(self, vertex):
        if vertex not in self.adj_list:
            self.adj_list[vertex] = {}

    def add_edge(self, vertex1, vertex2, weight):
        self.add_vertex(vertex1)
        self.add_vertex(vertex2)
        self.adj_list[vertex1][vertex2] = weight
        self.adj_list[vertex2][vertex1] = weight  # Uncomment for undirected graph

    def dijkstra(self, start_vertex):
        heap = [(0, start_vertex)]
        distances = {vertex: float('inf') for vertex in self.adj_list}
        distances[start_vertex] = 0

        while heap:
            current_distance, current_vertex = heapq.heappop(heap)

            if current_distance > distances[current_vertex]:
                continue

            for neighbor, weight in self.adj_list[current_vertex].items():
                distance = current_distance + weight

                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    heapq.heappush(heap, (distance, neighbor))

        return distances

#2)Implement Dijkstra's algorithm for finding the shortest path in a weighted graph.
# Creating a graph
graph = Graph()
graph.add_edge('A', 'B', 1)
graph.add_edge('A', 'C', 4)
graph.add_edge('B', 'C', 2)
graph.add_edge('B', 'D', 5)
graph.add_edge('C', 'D', 1)
graph.add_edge('D', 'E', 7)

# Running Dijkstra's algorithm starting from 'A'
start_vertex = 'A'
result = graph.dijkstra(start_vertex)

# Displaying the shortest distances from 'A' to each vertex
for vertex, distance in result.items():
    print(f"Shortest distance from {start_vertex} to {vertex}: {distance}")
```

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def is_identical(tree1, tree2):
    if not tree1 and not tree2:
        return True
    if not tree1 or not tree2:
        return False
    return (
        tree1.value == tree2.value
        and is_identical(tree1.left, tree2.left)
        and is_identical(tree1.right, tree2.right)
    )

def is_subtree(tree, subtree):
    if not tree and not subtree:
        return True
    if not tree or not subtree:
        return False

    if is_identical(tree, subtree):
        return True

    return is_subtree(tree.left, subtree) or is_subtree(tree.right, subtree)

# 2)Write a program to check if a tree is a subtree of another tree.
# Constructing the main tree
main_tree = TreeNode(3)
main_tree.left = TreeNode(4)
main_tree.right = TreeNode(5)
main_tree.left.left = TreeNode(1)
main_tree.left.right = TreeNode(2)

# Constructing the subtree
subtree = TreeNode(4)
subtree.left = TreeNode(1)
subtree.right = TreeNode(2)

# Checking if the subtree is in the main tree
result = is_subtree(main_tree, subtree)
print("Is subtree:", result)
# Output: Is subtree: True
```

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def tree_diameter(root):
    def height_and_diameter(node):
        if not node:
            return 0, 0

        left_height, left_diameter = height_and_diameter(node.left)
        right_height, right_diameter = height_and_diameter(node.right)

        current_height = max(left_height, right_height) + 1
        current_diameter = max(left_height + right_height, left_diameter, right_diameter)

        return current_height, current_diameter

    _, diameter = height_and_diameter(root)
    return diameter

# 5)Write a program to find the diameter of a tree.
# Constructing a binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.right = TreeNode(6)
root.left.left.left = TreeNode(7)
root.left.left.left.left = TreeNode(8)

# Finding the diameter of the tree
diameter = tree_diameter(root)
print("Diameter of the tree:", diameter)
```

```python
from collections import defaultdict

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
        self.time = 0

    def add_edge(self, u, v):
        self.graph[u].append(v)
        self.graph[v].append(u)

    def find_articulation_points(self):
        def articulation_points_dfs(u, visited, parent, ap, disc, low):
            children = 0
            visited[u] = True
            disc[u] = self.time
            low[u] = self.time
            self.time += 1

            for v in self.graph[u]:
                if not visited[v]:
                    children += 1
                    parent[v] = u
                    articulation_points_dfs(v, visited, parent, ap, disc, low)
                    low[u] = min(low[u], low[v])

                    # Check if u is an articulation point
                    if parent[u] == -1 and children > 1:
                        ap[u] = True
                    if parent[u] != -1 and low[v] >= disc[u]:
                        ap[u] = True

                elif v != parent[u]:
                    low[u] = min(low[u], disc[v])

        num_vertices = len(self.graph)
        visited = [False] * num_vertices
        disc = [float("inf")] * num_vertices
        low = [float("inf")] * num_vertices
        parent = [-1] * num_vertices
        ap = [False] * num_vertices

        for i in range(num_vertices):
            if not visited[i]:
                articulation_points_dfs(i, visited, parent, ap, disc, low)

        result = [i for i, is_ap in enumerate(ap) if is_ap]
        return result

# Example usage:
graph = Graph()
graph.add_edge(0, 1)
graph.add_edge(1, 2)
graph.add_edge(2, 0)
graph.add_edge(1, 3)
graph.add_edge(3, 4)
graph.add_edge(4, 5)
graph.add_edge(5, 3)

articulation_points = graph.find_articulation_points()
print("Articulation Points:", articulation_points)
```

```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_item(self, item):
        if item not in self.graph:
            self.graph[item] = {}

    def add_interaction(self, item1, item2, weight=1):
        self.add_item(item1)
        self.add_item(item2)

        self.graph[item1][item2] = weight
        self.graph[item2][item1] = weight  # Assuming interactions are symmetric

    def recommend_items(self, user_interactions, num_recommendations=3):
        recommendations = {}

        for item, weight in user_interactions.items():
            if item in self.graph:
                for neighbor, neighbor_weight in self.graph[item].items():
                    if neighbor not in user_interactions:
                        score = weight * neighbor_weight

                        if neighbor in recommendations:
                            recommendations[neighbor] += score
                        else:
                            recommendations[neighbor] = score

        # Sort recommendations by score in descending order
        sorted_recommendations = sorted(recommendations.items(), key=lambda x: x[1], reverse=True)

        return sorted_recommendations[:num_recommendations]

#2)Implement a simple recommendation system using a graph.
recommendation_graph = Graph()

# Adding items and interactions
recommendation_graph.add_interaction("MovieA", "MovieB", 0.8)
recommendation_graph.add_interaction("MovieA", "MovieC", 0.5)
recommendation_graph.add_interaction("MovieB", "MovieD", 0.6)
recommendation_graph.add_interaction("MovieC", "MovieD", 0.7)

# Simulating user interactions
user_interactions = {"MovieA": 1.0, "MovieB": 0.5}

# Getting recommendations for the user
recommendations = recommendation_graph.recommend_items(user_interactions, num_recommendations=2)

print("Recommendations:")
for item, score in recommendations:
    print(f"{item} (Score: {score})")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
from queue import Queue

def create_maze(dim):
    # Create a grid filled with walls
    maze = np.ones((dim*2+1, dim*2+1))

    # Define the starting point
    x, y = (0, 0)
    maze[2*x+1, 2*y+1] = 0

    # Initialize the stack with the starting point
    stack = [(x, y)]
    while len(stack) > 0:
        x, y = stack[-1]

        # Define possible directions
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        random.shuffle(directions)

        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if nx >= 0 and ny >= 0 and nx < dim and ny < dim and maze[2*nx+1, 2*ny+1] == 1:
                maze[2*nx+1, 2*ny+1] = 0
                maze[2*x+1+dx, 2*y+1+dy] = 0
                stack.append((nx, ny))
                break
        else:
            stack.pop()
            
    # Create an entrance and an exit
    maze[1, 0] = 0
    maze[-2, -1] = 0

    return maze
```

```python
def find_path(maze):
    # BFS algorithm to find the shortest path
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    start = (1, 1)
    end = (maze.shape[0]-2, maze.shape[1]-2)
    visited = np.zeros_like(maze, dtype=bool)
    visited[start] = True
    queue = Queue()
    queue.put((start, []))
    while not queue.empty():
        (node, path) = queue.get()
        for dx, dy in directions:
            next_node = (node[0]+dx, node[1]+dy)
            if (next_node == end):
                return path + [next_node]
            if (next_node[0] >= 0 and next_node[1] >= 0 and 
                next_node[0] < maze.shape[0] and next_node[1] < maze.shape[1] and 
                maze[next_node] == 0 and not visited[next_node]):
                visited[next_node] = True
                queue.put((next_node, path + [next_node]))
```

```python
def draw_maze(maze, path=None):
    fig, ax = plt.subplots(figsize=(10,10))
    
    # Set the border color to white
    fig.patch.set_edgecolor('white')
    fig.patch.set_linewidth(0)

    ax.imshow(maze, cmap=plt.cm.binary, interpolation='nearest')
    
    # Draw the solution path if it exists
    if path is not None:
        x_coords = [x[1] for x in path]
        y_coords = [y[0] for y in path]
        ax.plot(x_coords, y_coords, color='red', linewidth=2)
    
    ax.set_xticks([])
    ax.set_yticks([])
    
    # Draw entry and exit arrows
    ax.arrow(0, 1, .4, 0, fc='green', ec='green', head_width=0.3, head_length=0.3)
    ax.arrow(maze.shape[1] - 1, maze.shape[0]  - 2, 0.4, 0, fc='blue', ec='blue', head_width=0.3, head_length=0.3)
    
    plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

class SocialNetwork:
    def __init__(self):
        self.graph = nx.Graph()

    def add_person(self, person):
        if person not in self.graph:
            self.graph.add_node(person)
            print(f"{person} has joined the social network!")

    def add_friendship(self, person1, person2):
        self.add_person(person1)
        self.add_person(person2)
        if not self.graph.has_edge(person1, person2):
            self.graph.add_edge(person1, person2)
            print(f"{person1} and {person2} are now friends!")

    def are_friends(self, person1, person2):
        return self.graph.has_edge(person1, person2)

    def visualize_network(self):
        pos = nx.spring_layout(self.graph)  # Layout algorithm for visualization
        nx.draw(self.graph, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue')
        plt.show()

#4)Write a program that simulates social network connections using a graph.
social_network = SocialNetwork()

social_network.add_person("Alice")
social_network.add_person("Bob")
social_network.add_person("Charlie")

social_network.add_friendship("Alice", "Bob")
social_network.add_friendship("Bob", "Charlie")

print("Are Alice and Charlie friends?", social_network.are_friends("Alice", "Charlie"))

# Visualize the social network
social_network.visualize_network()
```

```python
import heapq

class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, node1, node2, weight):
        if node1 not in self.graph:
            self.graph[node1] = {}
        if node2 not in self.graph:
            self.graph[node2] = {}

        self.graph[node1][node2] = weight
        self.graph[node2][node1] = weight

def dijkstra(graph, start, end):
    heap = [(0, start)]
    visited = set()

    while heap:
        (current_cost, current_node) = heapq.heappop(heap)

        if current_node in visited:
            continue

        visited.add(current_node)

        if current_node == end:
            return current_cost

        for neighbor, weight in graph[current_node].items():
            if neighbor not in visited:
                heapq.heappush(heap, (current_cost + weight, neighbor))

    return float('inf')

#5)Implement a shortest-path algorithm in a graph to simulate GPS navigation.
road_network = Graph()

# Adding roads with distances
road_network.add_edge("A", "B", 5)
road_network.add_edge("B", "C", 3)
road_network.add_edge("C", "D", 7)
road_network.add_edge("D", "E", 2)
road_network.add_edge("A", "D", 10)

# Finding the shortest path
start_location = "A"
end_location = "E"
shortest_distance = dijkstra(road_network.graph, start_location, end_location)

if shortest_distance != float('inf'):
    print(f"The shortest distance from {start_location} to {end_location} is {shortest_distance} units.")
else:
    print(f"There is no path from {start_location} to {end_location}.")
```
